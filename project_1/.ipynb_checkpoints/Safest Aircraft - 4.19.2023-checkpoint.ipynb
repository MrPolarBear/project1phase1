```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('Data/df_share2.csv', index_col=0)
```

```python
df
```

```python
df.keys()
```

```python
df = df.drop(['Latitude','Longitude','Report.Status'], axis=1)
```

```python
df = df.drop(['perc_of_total_fatal', 'perc_of_total_serious', 'perc_of_total_minor','perc_of_total_uninjured','fatal_perc_pop' ], axis=1)
```

```python
df['total_passenger'] = df['Total.Fatal.Injuries'] + df['Total.Serious.Injuries'] + df['Total.Minor.Injuries'] + df['Total.Minor.Injuries'] + df['Total.Uninjured']
```

```python
df['Event.Date'] = pd.to_datetime(df['Event.Date'])
```

```python
df['Event_Year'] = df['Event.Date'].dt.year
```

```python
df = df[df['Event_Year'] >= 1985]
```

```python
df['Count'] = 1
```

```python
df.keys()
```

```python
Make_Eng = df.drop(['year', 'total_pop', 'Event_Year'], axis=1)
```

```python
Make_Eng = Make_Eng.groupby(['Make', 'Number.of.Engines']).sum() #.size() #.reset_index(name='Count')
```

```python
Make_Eng.head()
```

```python
Make_Eng['Perc_Un_Inj'] = (Make_Eng['Total.Uninjured'] / Make_Eng['total_passenger'])*100
```

```python
Make_Eng['Tot_Walk_Away'] = (Make_Eng['Total.Uninjured'] + Make_Eng['Total.Minor.Injuries'])
```

```python
Make_Eng['Perc_Walk_Away'] = ((Make_Eng['Total.Uninjured'] + Make_Eng['Total.Minor.Injuries'])  / Make_Eng['total_passenger'])*100
```

```python
Make_Eng = Make_Eng.reindex(columns=['Count', 'Total.Fatal.Injuries', 'Total.Serious.Injuries', 'Total.Minor.Injuries', 'Total.Uninjured', 'total_passenger', 'Perc_Un_Inj','Tot_Walk_Away', 'Perc_Walk_Away' ])
```

```python
Make_Eng['Perc_Un_Inj'] = Make_Eng['Perc_Un_Inj'].fillna(0)
```

```python
Make_Eng['Perc_Walk_Away'] = Make_Eng['Perc_Walk_Away'].fillna(0)
```
